## Create Machine Learning Models

```
from sklearn.model_selection import train_test_split
X = df.drop('DEATH_EVENT', axis = 1)
y = df['DEATH_EVENT']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Naive Bayes

```
from sklearn import naive_bayes
nb = naive_bayes.GaussianNB()
nb.fit(X_train, y_train)
y_pre=nb.predict(X_test)
```

```
from sklearn import metrics
# Print your accuracy score.
print('Accuracy score:'), print(metrics.accuracy_score(y_test, y_pre))
```

```
cm = metrics.confusion_matrix(y_test, y_pre)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=nb.classes_)
disp.plot()
```

34 True Negatives, 1 False Positives, 13 False Negatives, and 10 True Positives





## Decision Tree

```
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
decision_tree = DecisionTreeClassifier(random_state=0)
decision_tree.fit(X_train, y_train)
dt_pred = decision_tree.predict(X_test)
cm = metrics.confusion_matrix(y_test, dt_pred)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=nb.classes_)
disp.plot()
print('Accuracy score:')
print(metrics.accuracy_score(y_test, dt_pred))
```

## Logistic Regression

```
from sklearn.linear_model import LogisticRegression

ds=LogisticRegression()
ds.fit(X_train,y_train)
y_pred=ds.predict(X_test)
cm = metrics.confusion_matrix(y_test, y_pred)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=nb.classes_)
disp.plot()
print('Accuracy score:')
print(metrics.accuracy_score(y_test, y_pred))
```
